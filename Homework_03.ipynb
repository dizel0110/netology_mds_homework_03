```python
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
```

[PCA inverse transform manually](https://stackoverflow.com/questions/32750915/pca-inverse-transform-manually)

```python
def MSE(X, n_components):
    pca = PCA(n_components)
    pca.fit(X) 
    X_reduced = np.dot(X, pca.components_.T) # transform
    X_original = np.dot(X_reduced, pca.components_) # inverse_transform    
    return mean_squared_error(X, X_original)
```

```python
n_components = 65
for n_components in range (n_components):
    MSE_digits =  MSE(X, n_components)
    plt.scatter(n_components, MSE_digits, color='blue', s=40, marker='o')
```

Видно, что если уменьшать количество компонент до 20, то картинка восстановленная будет практически неизменной. Если продолжать уменьшать до 10, то картинка будет хуже. Есть некоторое значение, после которого будет ещё хуже (глобальный скачок по метрике MSE c 20 компонент). Примерно с 8 компонент картинка существенно ухудшается восстановленная, видно и по метрике.

```python
def draw_original(X, n_components):
    pca = PCA(n_components)
    pca.fit(X)
    X_reduced = np.dot(X, pca.components_.T) # transform
    X_original = np.dot(X_reduced, pca.components_) # inverse_transform
    plt.figure(figsize=(16, 6))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(X_original[i,:].reshape([8,8]));
```
